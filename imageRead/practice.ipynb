```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
img = cv2.imread('img.png')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGBA))
```

```python
arr = img[:,:,0]
arr1 = img[:,:,1]
arr2 = img[:,:,2]

plt.figure(figsize=(15, 6))

plt.subplot(1, 3, 1)
plt.title("blue channel graysacle image")
plt.imshow(arr, cmap='gray')


plt.subplot(1, 3, 2)
plt.title("green channel graysacle image")
plt.imshow(arr1, cmap='gray')

plt.subplot(1, 3, 3)
plt.title("red channel graysacle image")
plt.imshow(arr2, cmap='gray')
```

```python
arr = np.asarray(arr)
print(arr[:9, :9])
```

```python
plt.figure(figsize=(15, 3))
plt.subplot(1, 1, 1)
plt.title("original grascale image")
plt.imshow(arr, cmap='gray')

(width, height) = arr.shape
center = (width // 2, height // 2)
angle = 90  # Rotate by 30 degrees
scale = 1 #scale (zoom in when > 1 or zoom out when <1)
rotation_matrix = cv2.getRotationMatrix2D(center, angle, scale)
rotated_img = cv2.warpAffine(img, rotation_matrix, (width, height))

plt.subplot(1, 2, 2)
plt.title("rotated grascale image")
plt.imshow(rotated_img, cmap = 'gray')
```
